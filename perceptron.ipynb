```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
mean1 = np.array([-1, -2])
covar1 = np.array([[0.5, 0], [0, 0.5]])

points1 = np.random.multivariate_normal(mean1, covar1,100)
```

```python
mean2 = np.array([2, 1])
covar2 = np.array([[0.5, 0], [0, 0.5]])

points2 = np.random.multivariate_normal(mean2, covar2,100)
```

```python
D = []
for i in range(100):
  a = [points1[i], 1]
  b = [points2[i], -1]
  D.append(a)
  D.append(b)
```

```python
# Making sure that the data is seperable

for i in range(200):
  vector = D[i][0]
  label = D[i][1]
  if (label == 1):
    plt.scatter(vector[0], vector[1], color='red')
  else:
    plt.scatter(vector[0], vector[1], color='blue')
plt.show()
```

```python
MAX_EPOCHS = 1000
min_accuracy = 0.99
weights = np.zeros(3)
for epoch in range(MAX_EPOCHS):
    num_correct = 0
    for i in range(200):
        x, y = D[i]
        x_tilde = np.append(x, 1)
        y_hat = np.sign(np.dot(weights, x_tilde))
        if (y_hat != y):
            weights = weights + y * x_tilde
        else:
            num_correct += 1
    accuracy = num_correct / 200
    if (accuracy > min_accuracy):
        break
```

```python
weights, accuracy
```

```python
w0 = weights[2]
w1 = weights[0]
w2 = weights[1]
```

```python
# Plotting the data and seperating hyperplane
for i in range(200):
  vector = D[i][0]
  label = D[i][1]
  if (label == 1):
    plt.scatter(vector[0], vector[1], color='red')
  else:
    plt.scatter(vector[0], vector[1], color='blue')

x1 = np.linspace(-3, 3)
x2 = (-w0 - w1*x1) / w2
plt.plot(x1, x2, color='green')
plt.show()
```
